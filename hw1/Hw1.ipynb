```python
import numpy as np
import random
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
'''
QUESTION 7
'''
def set_axis_cartesian():
    """Setting an axes for plot viewing"""
    ax = plt.figure().add_subplot(1, 1, 1) # creates a plot in a 1 by 1 grid, in the 1st and only spot. Returns coordinate axes
    
    ax.spines['left'].set_position('center') # puts the left spine in the center, 
    ax.spines['bottom'].set_position('center') # bottom spine in the center
    
    #disables top and right spine so now we have just have x and y axis
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    ax.xaxis.set_ticks_position('bottom') # sets ticks at bottom spine (x axis)
    ax.yaxis.set_ticks_position('left') # set ticks at left spine (y axis)
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])

# two random points to define the line
p1 = np.random.uniform(-1, 1, size=2) # this is getting a random array of size 2 with each value being between -1 and 1
p2 = np.random.uniform(-1, 1, size=2) # an array of size 2 is a point

set_axis_cartesian()

plt.plot(p1[0], p1[1], 'go') # plotting p1 as a green o
plt.plot(p2[0], p2[1], 'mo') # plotting p2 as a magenta o

m = (p1[1] -p2[1])/(p1[0] - p2[0]) # m = (y1 - y2)/ (x1 -x2)
c = p1[1] - m*p1[0] # c = y - mx

x_axis = np.array([-1, 1])# graphing the line from x = [-1,1]
y_axis = m*(x_axis) + c # line equation
plt.plot(x_axis, y_axis, 'r', label='f(x)') # plot the line

# limit m and b to 3 decimal points
plt.title(f'Random Target Function f(x) = {m:.3f}x + {b:.3f}')
plt.legend(loc='upper left') # plot the legend

plt.show()

N = 10
x = np.random.uniform(-1, 1, size=(N,2)) # random 10 by 2 array, 10 random points  
```

```python
# ...and plot the points and the line

# plots each of the 10 points with a color
for i,randomPoint in enumerate(x):
    plt.plot( randomPoint[0], randomPoint[1], 'o', label=f'x[{i}]', c=np.random.rand(3,) )
    
# plots the line again
plt.plot(x_axis, y_axis, 'r', label='f(x)')

plt.title(f'Random Target Function f(x) = {m:.3f}x + {c:.3f}')
plt.legend(loc='upper left')

plt.show()
```

```python
# cx + dy = e
# y = mx + b
# mx - y = -b
a = m
b = -1
d = -c
# assume the correct classification is the line (this is the ground truth function y, also called the target function)
def y(x:np.ndarray):
    assert x.shape == (3,)
    return np.sign(a*x[1]+b*x[2]-d).astype(int)

## Perceptron model, this in combination with the weights, composes the hypothesis set, and yeilds the final hypothesis
def h(point:np.ndarray,weights):
    assert point.shape == (3,)
    # sign(wTx)
    #np.sign returns -1, 0, or 1
    return np.sign(np.dot( np.transpose(weights), point )).astype(int)

# this is introduced in index 0, because for the corresponding index in 0, specifies the threshold (kind of like intercept)
# wTx = w0 + w1x + w2y, the sign of this defines the model's classification of the point
xWithBias = np.insert(x, 0, 1, axis=1) # insert 1 into 1st index specifying the column(1st axis)

# Initialize weights at zero
w = np.zeros(3)
```

```python
'''# One run of PLA
QUESTION 7-8
'''
def converges():    
    for point in xWithBias:
            assert point.shape == (3,)
            #checking model against ground truth
            if h(point, w) != y(point):
                return False # there is still one that is incorrectly classified
    print(f'We converged after {counter} runs!')
    return True

counter = 0
while not converges():
#     clear_output(wait=True)
    counter += 1
    
    correctNs = [] # points correctly classified
    wrongNs = [] # points incorrectly classified
    
    # Split points into correctly and incorrectly classified
    for n,point in enumerate(xWithBias):
        if h(point, w) == y(point):
            correctNs.append(n) # appends the index of good point
        else:
            wrongNs.append(n) # index of bad point
            
    # Pick an incorrectly classified point at random
    updateN = random.choice(wrongNs)
    print(f'Updating for point {updateN}')
    
    # Update its weight vector
    print('Updating weight vector...')
    print(f'Original weights: {w}')
    for n,weight in enumerate(w):
        w[n] = w[n] + y(xWithBias[updateN]) * xWithBias[updateN][n] 
    print(f'\nUpdated weights: {w}')
    # w[0] + w[1]x + w[2]y <, >, or= 0
    set_axis_cartesian()

    mW = -w[1]/w[2]
    bW = -w[0]/w[2]

    x_axisW = np.array([-1, 1])# graphing the line from x = [-1,1]
    y_axisW = mW*(x_axisW) + bW # line equation

    plt.plot(x_axisW, y_axisW, 'r', label='g(x)') # plot the line
    
    x_axis = np.array([-1, 1])# graphing the line from x = [-1,1]
    y_axis = m*(x_axis) + b # line equation
    
    plt.plot(x_axis, y_axis, 'g', label='y(x)') # plot the line
    

    for point in xWithBias:
        if(y(point) > 0):
            plt.plot(point[1], point[2], 'o', color = 'red')
        else:
            plt.plot(point[1], point[2], 'o', color = 'blue')
    plt.plot(xWithBias[updateN][1], xWithBias[updateN][2], 'o', color = 'green')
    plt.legend(loc='upper left')
    plt.pause(0.1)
#     time.sleep(1)

plt.plot(x_axis, y_axis, 'g', label='y(x)') # plot the line
plt.plot(x_axisW, y_axisW, 'r', label='g(x)') # plot the line

for point in xWithBias:
    if(y(point) > 0):
        plt.plot(point[1], point[2], 'o', color = 'red')
    else:
        plt.plot(point[1], point[2], 'o', color = 'blue')
plt.plot(xWithBias[updateN][1], xWithBias[updateN][2], 'o', color = 'green')
print(converges())
```
